```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\R\Desktop\MASAI\PROJECT\A_cleaning_and_modification_for_pandas.csv", index_col = 0)
df.head()
```

```python
df['designation'] = df['designation'].replace({'(Drive)':"",'(Mumbai)':"",'(Bangalore, Gurgaon, )':"",'(Gujarat, Chandigarh, Indore)':"",'(Jamshedpur, Kolkata, Ranchi)':""}, regex=True)
df['designation'] = df['designation'].replace({'Sr.':"Senior"}, regex=True)
df.head()
```

```python
df['estab_year'] = df['estab_year'].replace({'Founded in':" "}, regex=True)
df.head()
```

```python
df['employees_count'] = df['employees_count'].replace({'More than 1000 employees':"1000+", ' employees':"",' ':""}, regex=True)
df.head()
```

```python
df['job_description'] = df['job_description'].replace({'Function: ':"", '→':":"}, regex=True)
df.head()
```

```python
df['skills'] = df['skills'].replace({'\\n':", "}, regex=True)
df.head()
```

```python
df['hr_name'] = df['hr_name'].replace({'\\n':"-"}, regex=True)
df.head()
```

```python
df['experience_req'] = df['experience_req'].replace({'0-0 Years':"Only Freshers"}, regex=True)
df.head()
```

```python
df.to_csv("B_cleaning_and_modification_for_excel.csv")
```
